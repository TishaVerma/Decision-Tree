```python
inputs=df.drop('salary',axis='columns')
targets=df['salary']
```

```python
lb_1=LabelEncoder()
lb_2=LabelEncoder()
lb_3=LabelEncoder()
```

```python
inputs['company_n']=lb_1.fit_transform(inputs['company'])
inputs['job_n']=lb_2.fit_transform(inputs['job'])
inputs['degree_n']=lb_3.fit_transform(inputs['degree'])
```

```python
inputs_n=inputs.drop(['company','job','degree'],axis='columns')
inputs_n
```

```python
model=tree.DecisionTreeClassifier()
```
